## What geographers have done

Geographers have been applied ML in their studies for a while. In satelite remote sensing study, either unsupervised classification or supervised classification have been used for identifying different subjects in a satelitte map. Because subjects like farmland, built-up area or water body have different textures or shape, or absorbs different band of light, these data are quite typical to be used in ML (see Jing, C, et al), especially when we could abtain a huge amount of data from the satellites nowadays. 

For human geographers, ML also have have been applied. On the internet, user contributed data are widely used in analysis, since the amount of data is abundant. By tracking the data from Twitter, geographer could even use ML tool to predict flu spreading (Vertalka, 2018)

## In genocide study

Saddly, ML is not common in genocide study yet. One reason of this might be, this study have a long quanlitative tradition, and the amount of data may not be strong enough to support ML study. In addition, ML, especially the neural network approaches, are more capable to do those analysis that we human being can easily tell but we do not know why, just like the classification of items on satelite images, where the reasoning process is like a black box. However, in quantative studies in genocide, what we are interested in, is the location, cause and the process of the killings, and the result of which is largely of uncertainty when we think about it. 

A potential aspect for us to use ML as the tool, is to find the locations of mass graves. In forensic sciences, the abnormalty of micro element in a certain location usually indicates the existance of graves (Schmitt, 2001). But there are two problems. First, most graves are not a result of a genocide, and the geographic location of a mass grave also have to follow certain micro and meso level pattern, subject to local landscape and enviornment. Second, also abnormalty could suggest the existance of mass graves, but because of the limitation of sampling density, usually the abnormality might not be significant enough to draw attention from researchers.

By combining data from different sources, we could create a larger dataset, and the minor abornamlty of soil elements coupling with the other social, landscape data, ML might be able to train a model for predicting the possibility of having a mass grave nearby a certain location.
